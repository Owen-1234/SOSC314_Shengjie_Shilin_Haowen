```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from pathlib import Path

desktop_path = Path.home() / "Desktop"
data_folder = desktop_path / "SOSC314" / "Data"
input_file = data_folder / "energy_narrative_W_cleaned.csv"

df = pd.read_csv(input_file)

reuters_text = " ".join(df[df['source'].str.contains('Reuters', case=False, na=False)]['cleaned_content'].astype(str))
guardian_text = " ".join(df[df['source'].str.contains('Guardian', case=False, na=False)]['cleaned_content'].astype(str))

wc_reuters = WordCloud(width=800, height=800, background_color='white', colormap='viridis', max_words=100).generate(reuters_text)
wc_guardian = WordCloud(width=800, height=800, background_color='white', colormap='magma', max_words=100).generate(guardian_text)

plt.figure(figsize=(20, 10))

plt.subplot(1, 2, 1)
plt.imshow(wc_reuters, interpolation='bilinear')
plt.title('Reuters Cloud', fontsize=20, fontweight='bold')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(wc_guardian, interpolation='bilinear')
plt.title('The Guardian Cloud', fontsize=20, fontweight='bold')
plt.axis('off')

plt.tight_layout(pad=0)
plt.savefig(data_folder / "western_media_wordclouds.png", dpi=300)
plt.show()
```
